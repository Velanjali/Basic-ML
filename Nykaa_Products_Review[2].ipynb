Using **for** loop,each categorical feature , a countplot will be displayed to show how the classes are distributed.The line 
**df.select_dtypes(include='object')** selects the columns with their values and displays them.Include an if-statement as to pick fewer classes using the line 
**.nunique()<10** .

```python
for column in prd.select_dtypes(include='object'):
  if prd[column].nunique() < 10:
    sns.countplot(prd['Product Reviews Count'])
    sns.set(rc={'figure.figsize':(300,6)})
    plt.show()
```

It returns the graph of count of Product Reviews Count. Output is the count of each number of reviews.

```python
sns.countplot(prd['Product Price'])
sns.set(rc={'figure.figsize':(500,6)})
```

It gives the output of how many products has the range of price.As it is the countplot() it counts of the Product price it gives how many products has same price range.



```python
sns.countplot(prd['Product Rating'])
sns.set(rc={'figure.figsize':(750,9)})
```

It gives the graph of count of Product rating.As it is the countplot() of rating of the products.So it gives the how many products have same rating.

```python
fig,ax=plt.subplots(figsize=(300,100))
sns.boxplot(x=prd['Product Reviews Count'],y=prd['Product Price'],ax=ax)
plt.show()
```

It gives the output of graph between Product reviews count and product price.We can observe that different different price products has different reviews.Mostly costly products has high reviews

Correlation analysis measurs statistical relationship between two different  variables.It is popular in field of predictive analysis.

```python
corrs = prd.corr()
print(corrs)
```

*It is used to find the pairwise correlation between the product reviews count and expected brand count.

Here is the map of correlation of both Product reviews count and Expected brand count.It is called Heat Map.

```python
plt.figure(figsize=(10,9))
sns.heatmap(corrs,cmap='GnBu_r',annot=True)
plt.show()                                              #heatmap of correlations.
```
